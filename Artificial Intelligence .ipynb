    import re
    import string
    def text_cleaner(text):
      if isinstance(text, str):
        text = text.lower()
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        text = re.sub(r'\s+', ' ', text)
        return text.strip()
    cleaner = lambda x: text_cleaner(x)

    df['Title'] = pd.DataFrame(df['Title'].apply(cleaner))
    df['PositiveReview'] = pd.DataFrame(df['PositiveReview'].apply(cleaner))
    df['NegativeReview'] = pd.DataFrame(df['NegativeReview'].apply(cleaner))
